```python
from pytube import YouTube
import os
import googleapiclient.discovery
```

```python
# -*- coding: utf-8 -*-

# Sample Python code for youtube.playlistItems.list
# See instructions for running these code samples locally:
# https://developers.google.com/explorer-help/code-samples#python


os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "YOUR GOOGLE API KEY"

youtube = googleapiclient.discovery.build(api_service_name, api_version, developerKey = DEVELOPER_KEY)
request = youtube.playlistItems().list(
        part="snippet",
        playlistId="RDEMx6c-FvtjqSqZSYoQqQZW3A"
        )

response = request.execute()

print(response)
```

```python
id_video_list = []

for i in range(len(response["items"])):
  id = response["items"][i]["snippet"]["resourceId"]["videoId"]
  id_video_list.append(id)

print(id_video_list)
```

```python
def download_audio_only(video_url, output_path):
    try:
        # Mendapatkan objek YouTube dari URL
        youtube = YouTube(video_url)

        # Memilih format audio dengan kualitas terbaik
        audio_stream = youtube.streams.filter(only_audio=True, file_extension='mp4').first()

        # Mendownload audio ke direktori yang ditentukan
        audio_stream.download(output_path)

        print("Download audio selesai!")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

for item in id_video_list:
  video_url = "https://www.youtube.com/watch?v={}".format(item)
  output_path = "save/audio"

  download_audio_only(video_url, output_path)
```
